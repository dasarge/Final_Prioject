```python
import pandas as pd
csv_file = "gold.csv"

#Read in File data and store into Pandas data frame
gold_df = pd.read_csv(csv_file)
gold_df.head(10)
```

```python
#Round decimals and change Date to datetime data type
gold_df = pd.DataFrame(gold_df.round(2))
gold_df['Date'] = pd.to_datetime(gold_df['Date'])

#Check data types to confirm that date = datetime & all others are floats
gold_df.dtypes
```

```python
#Drop NaNs
gold_df = gold_df.dropna()
```

```python
#Check that all NaNs were dropped
rows_with_nan = []
for index, row in gold_df.iterrows():
    is_nan_series = row.isnull()
    if is_nan_series.any():
        rows_with_nan.append(index)

print(rows_with_nan)
```

```python
#Set date as index

gold_df = gold_df.set_index('Date')
gold_df.head(20)
```

```python
#Introduce new bitcoin dataframe of only the dates desired
new_gold_df = pd.DataFrame(gold_df.loc['2018-04-01':'2021-04-01'])
new_gold_df
```

```python
#Export to CSV
new_gold_df.to_csv('clean_gold.csv')
```
