```python
import pandas as pd
csv_file = "bitcoin.csv"

#Read in File data and store into Pandas data frame
bitcoin_df = pd.read_csv(csv_file)
bitcoin_df.head(10)
```

```python
#Round decimals and change Date to datetime data type
bitcoin_df = pd.DataFrame(bitcoin_df.round(2))
bitcoin_df['Date'] = pd.to_datetime(bitcoin_df['Date'])

#Check data types to confirm that date = datetime & all others are floats
bitcoin_df.dtypes
```

```python
#Drop NaNs
bitcoin_df = bitcoin_df.dropna()
```

```python
#Check that all NaNs were dropped
rows_with_nan = []
for index, row in bitcoin_df.iterrows():
    is_nan_series = row.isnull()
    if is_nan_series.any():
        rows_with_nan.append(index)

print(rows_with_nan)
```

```python
#Set date as index

bitcoin_df = bitcoin_df.set_index('Date')
bitcoin_df.head(20)
```

```python
#Introduce new bitcoin dataframe of only the dates desired
new_bitcoin_df = pd.DataFrame(bitcoin_df.loc['2018-04-01':'2021-04-01'])
new_bitcoin_df
```

```python
#Export to CSV
new_bitcoin_df.to_csv('clean_bitcoin.csv')
```
