```python
import pandas, sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
clean_fear = pd.read_csv("c:/Users/George Aziz/desktop/Final_Project/Clean_CSVs/clean_Fear-and-Greed-Sentiment.csv")



clean_fear.head(20)
```

```python
# Filter the DataFrame down only to those columns to chart
fear_and_greed = clean_fear[["Sentiment","Fear and Greed Index"]]

# Set the index to be "State" so they will be used as labels
fear_and_greed = fear_and_greed.set_index("Sentiment")

fear_and_greed.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
fear_and_greed.plot(kind="bar", figsize=(20,3))

# Set a title for the chart
plt.title("fear and greed index")

plt.show()
plt.tight_layout()
```
